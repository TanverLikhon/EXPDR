```python
import numpy as np
import pandas as pd
from sklearn import datasets 
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Chinese database.csv') 
df
```

```python
df.shape
```

```python
X = df.iloc[:, 0:79].values 
y = df.iloc[:, 79].values 
```

```python
#np.unique(y)
ydf=df.iloc[:,79]
ydf
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf
```

```python
finalDf = pd.concat([principalDf, ydf], axis = 1)
finalDf
```

```python
x=finalDf['principal component 1']
y=finalDf['principal component 2']
```

```python
fig,ax=plt.subplots()
for color in ['tab:blue','tab:red']:
    ax.scatter(x,y,c=color,label=color)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

targets = ['1', '0']
#colors = ['r', 'g']


#df.loc[df['B']==3, 'A'].item()
for color in ['tab:blue','tab:red']:
    ax.scatter(finalDf['principal component 1'],finalDf['principal component 2'],c=color,label=color)


```

```python
X=principalDf
y=ydf
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) 
```

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler() 

X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test) 
```

```python
from sklearn.linear_model import LogisticRegression 

classifier = LogisticRegression(random_state = 0) 
classifier.fit(X_train, y_train) 
classifier.score(X_test,y_test)
```

```python
y_pred=classifier.predict(X_test)
```

```python
# making confusion matrix between 
# test set of Y and predicted value. 
from sklearn.metrics import confusion_matrix 
#precision recall F1 score
cm = confusion_matrix(y_test, y_pred) 
cm
```

```python
# Predicting the training set 
# result through scatter plot 
from matplotlib.colors import ListedColormap 

X_set, y_set = X_train, y_train 
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
					stop = X_set[:, 0].max() + 1, step = 0.01), 
					np.arange(start = X_set[:, 1].min() - 1, 
					stop = X_set[:, 1].max() + 1, step = 0.01)) 



plt.xlim(X1.min(), X1.max()) 
plt.ylim(X2.min(), X2.max()) 

for i, j in enumerate(np.unique(y_set)): 
	plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], 
				c = ListedColormap(('red', 'green'))(i), label = j) 

plt.title('Logistic Regression (Training set)') 
plt.xlabel('PC1') # for Xlabel 
plt.ylabel('PC2') # for Ylabel 
plt.legend() # to show legend 

# show scatter plot 
plt.show() 
```
